```python
# Import the required packages
import pandas as pd
from pytrends import dailydata
```

```python
# Daily trends data for "bitcoin" from Jan 2015 to Sep 2022

"""
get_daily_data(word: str,
                 start_year: int,
                 start_mon: int,
                 stop_year: int,
                 stop_mon: int,
                 geo: str = 'US',
                 verbose: bool = True,
                 wait_time: float = 5.0)
"""
daily_df = dailydata.get_daily_data('bitcoin', 2015, 1, 2022, 9)
```

```python
daily_df.drop(['isPartial'],axis=1)
print(daily_df)

"""
bitcoin_unscaled -> original daily data fetched month by month, and it is not comparable across different months
bitcoin_monthly -> original monthly data fetched at once
bitcoin -> daily search volume already scaled and comparable through time
scale -> scale used to obtain the scaled daily data

"""
```

```python
# Format the dataframe by removing index and formatting the date column in Y-M-D format
daily_df['date'] = daily_df.index
daily_df.reset_index(drop=True,inplace=True)
daily_df['date'] = pd.to_datetime(daily_df['date']).dt.date
formatted_daily_df = daily_df[['date','bitcoin','bitcoin_unscaled','bitcoin_monthly','scale']]
formatted_daily_df
```

```python
# Save the results in a csv file
formatted_daily_df.to_csv("Google_Trends_Bitcoin_Daily_Data.csv",index=False)
```
